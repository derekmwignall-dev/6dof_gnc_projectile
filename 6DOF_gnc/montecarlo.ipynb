# 6 DOF Ballistics Simulation Visualization
Visualizes trajectory and orientation from the C++ simulation output.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib widget

# Load simulation data
df = pd.read_csv('montecarlo.csv')
run  = df['run'].values
x  = df['dx'].values
y  = df['dy'].values
z  = df['dz'].values
vx = df['dvx'].values
vy = df['dvy'].values
vz = df['dvz'].values
missDistance = df['missDistance'].values

missDistance = np.where(missDistance <= 10, 10, 11)
        
interceptTime = df['interceptTime'].values


print(f'Loaded {len(df)} runs')
```

## 1. Initial Position vs. Miss

```python
fig = plt.figure(figsize=(12, 8))
ax  = fig.add_subplot(111)
magnitude = np.sqrt(x**2+y**2+z**2)

# Plot trajectory
ax.scatter(magnitude, missDistance)

# ax.set_xlim((0,1200))
ax.set_ylim((0,20))
ax.legend()
plt.tight_layout()
ax.set_xlabel('Initial Missile-Target Distance (ft)')
ax.set_ylabel('Miss Distance (ft)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate magnitudes
distance = np.sqrt(x**2 + y**2 + z**2)
velocity = np.sqrt(vx**2 + vy**2 + vz**2)

# Plot
fig, ax = plt.subplots(figsize=(9, 6))

scatter = ax.scatter(distance, velocity, c=np.log(missDistance), cmap='plasma', alpha=0.7, edgecolors='none')

plt.colorbar(scatter, ax=ax, label='Miss Distance')

ax.set_xlabel('Distance')
ax.set_ylabel('Velocity')
ax.set_title('Distance vs Velocity (colored by Miss Distance)')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax  = fig.add_subplot(111)

# Plot trajectory
ax.scatter(vz, missDistance)

# ax.set_xlim((0,1200))
# ax.set_ylim((0,20))
ax.legend()
plt.tight_layout()
ax.set_xlabel('Initial Missile-Target Distance (ft)')
ax.set_ylabel('Miss Distance (ft)')
plt.show()
```
